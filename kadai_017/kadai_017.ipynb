```python
class Human:
   def __init__(self, name, age):
       self.name = name
       self.age = age
   def check_adult(self):
       if self.age >= 20:
        print(f"{self.name}は大人である")
       else:
        print(f"{self.name}は大人でない")

user1=Human("太郎",30)
user2=Human("花子",15)
user3=Human("次郎",20)

user_names=[user1,user2,user3]

for user in user_names:
  user.check_adult()
```
